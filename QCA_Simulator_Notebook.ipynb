```python
# Paso 1: Definir la clase QCASimulator
import numpy as np
import matplotlib.pyplot as plt

class QCASimulator:
    def __init__(self, time_window=25, resolution=0.1):
        self.time_window = time_window
        self.resolution = resolution
        self.timeline = np.arange(0, time_window, resolution)
        self.data = {}

    def simulate(self):
        self.data["State Prep"] = np.exp(-((self.timeline - 2.5) ** 2) / (2 * 0.8**2))
        self.data["Affinity Calc"] = np.exp(-((self.timeline - 10) ** 2) / (2 * 1.5**2))
        self.data["Amplification"] = np.exp(-((self.timeline - 19) ** 2) / (2 * 1.2**2))
        self.data["Transduction"] = np.exp(-((self.timeline - 24) ** 2) / (2 * 0.5**2))

    def plot(self):
        plt.figure(figsize=(10, 4))
        for label, signal in self.data.items():
            plt.plot(self.timeline, signal, label=label)
        plt.title("Quantum Consciousness Processing Timeline")
        plt.xlabel("Time (ms)")
        plt.ylabel("Activation Level")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()


# Paso 2: Ejecutar simulación y mostrar gráfica
sim = QCASimulator()
sim.simulate()
sim.plot()
```
